```python
#infix to postfix
class Node:
    def __init__(self,elem,nxt):
        self.elem=elem
        self.next=nxt
class Stack:
    def __init__(self):
        self.top=None
    def push(self,elem):
        if self.top==None:
            self.top=Node(elem,None)
        else:
            n=Node(elem,self.top)
            self.top=n
    def pop(self):
        if self.top==None:
            return "Underflow"
        else:
            temp=self.top.elem
            self.top=self.top.next
            return temp
        
    def peek(self):
        if self.top==None:
            return "Underflow"
        else:
            return self.top.elem    
        
        
        
s=Stack()
giv="K+L-M*N+(O^P)*W/U/N*T+Q"        
final=''
for i in giv:
    if ord("A")<=ord(i)<=ord("Z") or ord("a")<=ord(i)<=ord("z"):
        final+=i
    elif i in "+-*^/()": #operators and parenthesis
        if s.peek()=="Underflow": #for all
            s.push(i)
            print(s.peek())
        else:
            if i in "+-*/^": #for all
                if s.peek()=="(" or s.peek()=="{" or s.peek()=="[" :
                    push(i)
                    print(i)
                elif i in "+-": #for +-
                    if s.peek() in "^/*+-":
                        while s.peek() not in "({[" or s.peek()!="Underflow":
                            val=s.pop()
                            if val not in "({[" or val!="Underflow":
                                final+=val
                                        
                elif i in "/*^": #for */^
                    if s.peek() in "+-":
                        push(i)
                elif i in ")}]": #for close brackets
                    while s.peek() not in "({[":
                        val=s.pop()
                        if val not in "({[" or val!="Underflow":
                            final+=val
if s.peek()!="Underflow":
    while s.peek()=="Underflow":
        val=s.pop()
        if val!="Underflow":
            final+=val
print(final)
```
