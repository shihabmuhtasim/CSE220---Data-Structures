```python
#2
def RemoveRightShift(cir,start,size,pos):
    to= (start)
    fr= (to+1)%len(cir)
    nshift= (size-pos+1)
    last=(start+size-1)%len(cir)
    for i in range(nshift):
        cir[to]=cir[fr]
        to=fr
        fr=(to+1)%len(cir)
    cir[last]=0
    return cir

def removePositive (cir, start, size):
    k=start
    pos=0
    while pos!=size-1:
        if cir[k]>0:
            cir=RemoveRightShift(cir,k,size,pos)
            #print(cir, "start",k, "size",size,"offset" ,pos)
            size-=1
        else:
            pos=(pos+1)
            k=(k+1)%len(cir)
        
    print(cir)
    
removePositive([4,-5,-6,0,0,1,-2,-3],5,6)
```

```python
#2problem 
class Node:
    def __init__(self,elem,next):
        self.elem=elem
        self.next=next
class Linkedlist:
    def __init__(self,a):
        self.head=Node(a[0],None)
        temp=self.head
        for i in range (1,len(a)):
            n=Node(a[i],None)
            temp.next=n
            temp=n
    def print(self,head):
        temp=head
        while temp!=None:
            print(temp.elem)
            temp=temp.next
        print()
def Reverse(head,k):
    current = head 
    next = None
    prev = None
    count = 0
  
    #1) reverse first k nodes of the linked list 
    while (current != None and count < k) : 
        next = current.next
        current.next = prev 
        prev = current 
        current = next
        count = count + 1
      
    # 2) Now head pos to the kth node. 
    # So change next of head to (k+1)th node
    if(head != None): 
        head.next = current 
  
    # 3) We do not want to reverse next k 
    # nodes. So move the current 
    # pointer to skip next k nodes 
    count = 0
    while(count < k - 1 and current != None ): 
        current = current.next
        count = count + 1
      
    # 4) Recursively call for the list 
    # starting from current.next. And make
    # rest of the list as next of first node 
    if(current != None): 
        current.next = Reverse(current.next, k) 
  
    # 5) prev is new head of the input list 
    return prev 
        
        
    
    
        

a=[1,2,3,4,5,6,7,8]
s=Linkedlist(a)
x=Reverse(s.head,4)
print(x)
```
